# Rede Neural Artificial - Classificação Multiclasses

```python
import pandas as pd
import numpy as np
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import models, layers
```

```python
tf.__version__
```

## Obter os dados

```python
data = pd.read_csv("iris.csv")
```

```python
data.head(1)
```

## Transformação de variaveis categóricas em quantitativas 

```python
# https://hackernoon.com/what-is-one-hot-encoding-why-and-when-do-you-have-to-use-it-e3c6186d008f
# https://machinelearningmastery.com/why-one-hot-encode-data-in-machine-learning/

# https://www.quora.com/Minmaxscaler-vs-Standardscaler-Are-there-any-specific-rules-to-use-one-over-the-other-for-a-particular-application
# https://datascience.stackexchange.com/questions/45900/when-to-use-standard-scaler-and-when-normalizer
# https://jovianlin.io/feature-scaling/

# Não se utiliza o label encoder(todas as classes num vector[150,1]), porque cada classe tem de ter a probabilidade de pertencer
# a cada uma das flores, ou seja tem de ser uma matriz [150,3] (exigencia da biblioteca).

# iris setosa:    1 0 0
# iris verginica  0 1 0
# iris versicolor 0 0 1
class_dummy = pd.get_dummies(data["class"])
```

```python
class_dummy.head()
```

## Divisão Treino-Teste 

```python
# X - variáveis usadas para modelar o modelo
X = data.drop("class", axis=1)
X.head(1)
```

```python
# y - variavel que se pretende avaliar
y = class_dummy
y.head(1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

## Modelo Artificial Neural Network

```python
# sequential - uma cada apos a outra
# dense - camadas fully conected, cada neuronio esta conectado a todos os neuronios da camada seguinte

# 1ª Camada Oculta(e definição da camada de entrada) 
# Neuronios: normalmente escolhe-se o número de neuronios igual ao número de variaveis de entrada mais o número de variaveis de 
# saida divididos por 2 ((4+3)/2=3.5=4) para começar a modelar a rede. 
# Função de Activação: Normalmente escolhe-se a reLu para deep learning porque obtem-se melhores resultados do que com uma
# função sigmoide ou tangente hiperbolica. 
# input_dim: quantos atributos existem na camada de entrada (neste caso são 4 variaveis de entrada). Este parâmetro só é 
# necessário para a primeira camada oculta.

# 2ª Camada Oculta
# Na segunda camada oculta em geral coloca-se a mesma quantidade de neurónios do que na primeira.

# Camada de Saida
# Neuronios: Igual ao número de saidas possiveis, neste caso são 3 classes.
# Função de activação: softmax é a função utilizada para problemas de classificação multiclasse. Para se obter a probabilidade
# para cada uma das classes (por ex. para um neurónio tem-se a probabilidade de 10% de ser iris setosa, 80% iris versicolor e
# 10% de ser iris verginica. A classe atribuida será aquela que tem maior probabilidade, neste ex. a iris verginica).

model = models.Sequential()
model.add(layers.Dense(units=4, activation="relu", input_dim=4))
model.add(layers.Dense(units=4, activation="relu"))
model.add(layers.Dense(units=3, activation="softmax"))
```

## Compilação e Ajuste do modelo ANN

```python
# Optimizer: adam é uma optimização da descida do gradiente estócastico (usado para encontrar os melhores valores dos pesos).
# É o que melhor se adapta à maioria dos casos.

# loss = função de perda categorical_crossentropy é a mais utilizada para classificação multiclasse.


model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["categorical_accuracy"])
```

```python
# batch_size: 10, calcula-se o erro de um conjunto de 10 dados e actualizam-se todos os pesos. 

# epochs: Quantas vezes o algoritmo percorre o conjunto de dados de treino totalmente.

# Por exemplo para um conjunto de dados de treino de 436 registros e batch_size de 6 registros:
# 426/6 = 71 iterações por época. 71 actualizações dos pesos por época

# Ajustar os dados de entrada de treino aos dados de saida de treino para treinar o modelo
model.fit(X_train, y_train, batch_size=10, epochs=100)
```

```python
# A avalição da binary_accuracy: 0.97 está a ser feita no conjunto de treino, ou seja, está a fazer o treino e a avaliar no
# mesmo conjunto de dados. A avaliação deve ser feita para um conjunto de dados novos(não treinados).
```

## Previsão e Avaliação do Modelo ANN

```python
# Prever os dados de teste para avaliar o modelo
predictions = model.predict(X_test)
predictions[0:3]
```

```python
# de modo a se poder utilizar a matriz de confusão e o relatorio de classificação para avaliar o modelo tem-se de transformar 
# o vectores prediction e y_teste do formato [150,3] para [150,1] (exigencia da biblioteca sklearn).

# Retorna o número do indice que tem a probabilidade mais alta.
predictions_index = [np.argmax(index) for index in predictions]
predictions_index[0:4]
```

```python
# Retorna o número do indice que tem a probabilidade mais alta.
y_test_index = [np.argmax(index) for index in np.array(y_test)]
y_test_index[0:4]
```

```python
pd.DataFrame({"prediction index": predictions_index, "teste index":y_test_index}).head(5)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#Relatório de classificação
#Ver como o modelo se ajusta para os dados de teste
print(classification_report(y_test_index, predictions_index))
```

```python
confusion = confusion_matrix(y_test_index,predictions_index)
```

```python
#Matriz de confusão
outside_columns = ["","Predicted",""]
outside_index = ["","Actual",""]
inside = ["iris setosa","iris verginica","iris versicolor"]

hier_columns = list(zip(outside_columns, inside))
hier_columns = pd.MultiIndex.from_tuples(hier_columns)

hier_index = list(zip(outside_index, inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)

pd.DataFrame(confusion, columns=hier_columns, index = hier_index)
```

```python
# Com o TensorFlow 2.0 pode-se avaliar o modelo de forma mais expedita através do método evaluate. 
# Sem ser necessário criar o relatorio de classificação e as respectivas transformações na variavel de saida (transformar o 
# array de [150,3] para [150,1].
# Os métodos de avaliação podem ser definidos quando se cria o modelo (neste caso escolheu-se apenas a accuracy).
model.evaluate(X_test,y_test)
```

## Validação Cruzada

```python
# Utiliza-se o KerasClassifier porque é um problema de classificação

from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_validate
```

### Transformação de variaveis categóricas em quantitativas 

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
# Para a cross validation não se pode utilizar o pd.gotDummies() porque ocorre um erro por causa das dimensões do array.
# Neste caso tem de se utilizar um array [150,1] e não [150,3] (exigencia da biblioteca sklearn).

y_enc = enc.fit_transform(data['class'])
y_enc
```

### Criar o Modelo

```python
### Criar o modelo de DNN numa função

def model():
    model = models.Sequential()
    model.add(layers.Dense(units=4, activation="relu", input_dim=4))
    model.add(layers.Dense(units=4, activation="relu"))
    model.add(layers.Dense(units=3, activation="softmax"))
    model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["categorical_accuracy"])
    return model
```

```python
model = KerasClassifier(build_fn=model, epochs=1000, batch_size=10)
```

### K-Fold Cross Validation

```python
# cross validation

scoring = ['precision_weighted', 'recall_weighted', 'f1_weighted']

scores = cross_validate(estimator=model, X=X, y=y_enc, cv=4, scoring="accuracy", return_train_score=False)
```

```python
# Valores obtidos para a accuracy atraves de cross validation
scores
```

```python
# accuracy média dos sub conjuntos de teste
np.mean(scores["test_score"])
```

```python
# desvio padrão obtido atraves dos valores da accuracy dos sub conjuntos de teste
# quanto maior este valor maior é a probabilidade de existir overfitting (o modelo não generaliza bem e adapta-se demasiado bem
# aos dados de treino, e quando se utilizam dados novos o modelo não se adapta bem)

np.std(scores["test_score"])
```

## Ajuste de parâmetros

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Criar o modelo de DNN numa função, com dropout e ajuste de parâmetros.
# Tem de se usar a sparse_categorical_crossentropy para a função perda.
# Tem de se usar a accuracy como parametro.

def model(optimizer, activation, neurons):
    model = models.Sequential()
    model.add(layers.Dense(units=neurons, activation=activation, input_dim=4))
    model.add(layers.Dropout(0.2)) 
    model.add(layers.Dense(units=neurons, activation=activation))
    model.add(layers.Dropout(0.2)) 
    model.add(layers.Dense(units=3, activation="softmax"))
    
    model.compile(optimizer=optimizer, loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    return model
```

```python
model = KerasClassifier(build_fn=model )
```

```python
parametros = {'batch_size': [10, 30],
              'epochs': [50, 100],
              'optimizer': ['adam'],
              'activation': ['relu', 'tanh'],
              'neurons': [8, 4]}
```

```python
#  Construção da grelha
# Não se pode utilizar o parametro scoring na classificação multiclasses (exigencia da biblioteca sklearn).
grid = GridSearchCV(estimator=model, param_grid=parametros, cv=2, refit=True)
```

```python
# Para a GridSearchCV não se pode utilizar o pd.gotDummies() porque ocorre um erro por causa das dimensões do array.
# Neste caso tem de se utilizar um array [150,1] e não [150,3] (exigencia da biblioteca sklearn).

# Ajustar o modelo com os parametros escolhidos e para o conjunto de dados total(a cross validation é feita pelo GridSearch).
grid.fit(X,y_enc)
```

```python
# Os melhores parametros para o modelo
grid.best_params_
```

## Gravar a Rede Neural

```python
# gravar a rede neural final(já treinada com os melhores parâmetros, pesos e estrutura) para não ser necessário fazer de novo 
# o treino. Assim já não é necessário fazer o treino quando se quer prever um novo registro.
```

### Criar o modelo final

```python
# Modelo final
# 8 neurónios e activation="tanh" porque foram os parametros encontrados atraves da gridsearch que optimizam o modelo.

model = models.Sequential()
model.add(layers.Dense(units=8, activation="tanh", input_dim=4))
model.add(layers.Dropout(0.2)) 
model.add(layers.Dense(units=8, activation="tanh"))
model.add(layers.Dropout(0.2)) 
model.add(layers.Dense(units=3, activation="softmax"))   
```

### Compilação e Ajuste do modelo final

```python
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["categorical_accuracy"])
```

```python
# Podem-se utilizar os dados todos porque o modelo já foi avaliado anteriormente
model.fit(X, y, epochs=100, batch_size=10)
```

```python
model_json = model.to_json()
```

```python
# gravar o modelo no formato json
with open('model_iris.json', 'w') as json_file:
    json_file.write(model_json)
```

```python
# gravar os pesos no formato h5
model.save_weights('weights_iris.h5')
```
