# Redes Adversariais Generativas(GANs) - Geração Automática de Imagens

```python
# Rede neural que pode criar dados. Aprende com os dados e cria novos dados(outras versões dos dados aprendidos que não 
# existiam anteriormente).

# Pode ser utilizada para:
# Criar textos. Por exemplo com base nos textos de um poeta pode criar poesia com as caracteristicas desse poeta.
# Aumentar a resolução de imagens.
# Desenhar automáticamente.
# Converter texto em imagem.
# Converter imagem em imagem. Por exemplo tem-se a visão noturna de uma fachada e cria-se a imagem real da fachada. Passar
# a imagem de preto e branco para cores. Tem-se as bordas da imagem tracejadas e cria-se a imagem completa.
# Geração de automáticas de imagens. Google Deep Mind.
# Criação de Ambientes. Por exemplo gerar várias opções de design interior para uma divisão da casa.

# Aprendizagem de criação de imagens.
# A GAN possui 2 componentes:
# O gerador que cria as imagens e o discriminador que acessa as imagens criadas e informa se elas são parecidas às originais.
# As 2 componenetes Começam do zero, ou seja não sabem nada sobre as imagens, e aprendem sozinhos. O gerador aprende a criar
# imagens e o discriminador aprende a avaliar as imagens. Ou seja tem-se 2 redes neurais.
# 1 rede neural para aprender a criar imagens e outra rede neural para apender a avaliar as imagens.
# O gerador tem como entrada um vector com números aleatorios e gera uma imagem com base nesse vector. Com o avançar das épocas
# a rede neural vai aprendendo(melhorando os pesos) e vai gerar imagens mais parecidas às imagens pretendidas.
# O discriminador recebe como parametro as imagens pretendidas e recebe tambem imagens diferentes das pretendidas. Tem-se um
# problema de classificação binária(imagens pretendidas e imagens não pretendidas).
# Tanto se pode utilizar redes neurais densas como redes neurais convolucionais, que são mais especificas para imagens.

# 1 itapa:
# Gerador: recebe numeros aleatorios e gera imagens. Como a rede começa a aprender do zero estas primeiras imagens são
# muito diferentes da pretendida. As imagens são constituidas pelo conjunto de pixeis gerado pela rede(A imagem é um conjunto
# de pixeis que variam entre 0 e 255 ou seja a imagem irá ser uma matriz de números).

# Discriminador: As primeiras imagens geradas são passadas para o discriminador e tambem a base de dados com as imagens
# pretendidas é passada para o discriminador. Tem-se um problema de classificação binária(imagens pretendidas e imagens não
# pretendidas), com 1 neuronio na camada de saida e função de activação sigmoid.
# Nesta época o discriminador ainda não foi treinado e não sabe ainda quais são as imagens pretendidas. Vai retornar a 
# probabilidade das imagens geradas serem as não pretendidas(o esperado é que a saida seja 0). E vais retornar a probabilidade
# das imagens da base de dados com as imagens pretendidas(o esperado é que a saida seja 1).
# Por exemplo se tiver-se 500 imagens de cães(da base de dados) e 500 imagens de não cães(do gerador) a rede neural vai fazer
# o treino durante várias épocas utilizado o algoritmo de back propagation para calculo do erro e vai actualiar os pesos. 
# Nesta etapa o discriminador vai aprender o que é cão e o que não é. Depois do treino feito o discriminador já não necessita 
# de ter como entrada a base de dados dos cães porque a rede neural já foi treinada.
# Volta-se a passar as imagens do gerador(já sem a base de dados dos cães) para o classficador(já treinado) fazer a 
# classificação dessas imagens(quer-se que a saida aproxime-se de 1). Por exemplo tem-se a seguinte classificação(0.3 0.6 0.3).
# Compara-se a classificação com o valor 1 e calcula-se o erro.
# A classificação obtida no discriminador é enviada para o gerador. Está-se a indicar ao gerador que as imagens ainda não estão
# proximas das pretendidas(têm erro). Ou seja o gerador passa as imagens geradas para o discriminador e o discriminar classifica
# essas imagens é diz que o erro ainda é demasiado alto. Indica que é nescessário melhorar as imagens(por exemplo o cão precisa
# de orelhas, rabo, etc.). Isso é feito pelo calculo do erro.
# Depois do discriminador passar o feedback para o gerador o processo é iniciado novamente. 

# 2 itapa:
# Gerador: recebe numeros aleatorios e gera imagens. Estas imagens já vão ser mais semelhantes às pretendidas.
# Discriminador: As imagens geradas são passadas para o discriminador e tambem a base de dados com as imagens pretendidas é 
# passada para o discriminador. O discriminador vai retornar a probabilidade das imagens gerdas serem as não pretendidas(o 
# esperado é que a saida seja 0). E vais retornar a probabilidade das imagens da base de dados com as imagens pretendidas(o 
# esperado é que a saida seja 1). A rede neural discriminador vai fazer o treino durante várias épocas utilizado o algoritmo de 
# back propagation para calculo do erro e vai actualiar os pesos. Nesta etapa o discriminador vai aprender o que é cão e o que 
# não é. Depois do treino feito o discriminador já não necessita de ter como entrada a base de dados dos cães porque a rede 
# neural já foi treinada.
# Volta-se a passar as imagens do gerador(já sem a base de dados dos cães) para o classficador(já treinado) fazer a 
# classificação dessas imagens(quer-se que a saida aproxime-se de 1). Por exemplo tem-se a seguinte classificação(0.6 0.8 0.5).
# Compara-se a classificação com o valor 1(imagens de cães) e calcula-se o erro.
# A classificação obtida no discriminador é enviada para o gerador.

# O discriminador v ai dar o feedback da classificação para o gerador. A rede neural geradora vai pegar no erro obtido na
# classificação e fazer o treino da rede e gerar novas imagens de modo a melhorar esse erro. O feedback pode ser por exemplo
# que as imagens têm olhos a mais, bocas a mais, etc. Este processo é repetido até se obterem imagens semelhantes às 
# pretendidas.

# Instalação do Keras Adversarial.
# https://github.com/bstriner/keras-adversarial
```

# Pré-Processamento dos dados

### Conversão dos dados do tipo int8 para float32

```python
# Como os dados vão ser convertidos para uma escala entre 0 e 1 tem-se de converter os dados de inteiro8 para float32 para que
# os valores obtidos depois da normalização não sejam inteiros e consequentemente quase todos 0(por não haver números decimais).

type(X_train[0,0,1])
```

```python
# Conversão dos dados em float 32
X_train = X_train.astype("float32")
```

## Modelo do Gerador(Rede Neural Arificial)

```python
# Gerador

# 1ª camada oculta e definição da camada de entrada: 500 neuronios na camada oculta(obtido pelo afinamento de parametros).
# input_dim: Nº de neuronios na camada de entrada: 100 imagens(são geradas 100 novas imagens).
# 2ª camada oculta
# Camada de saida: units=vector com 784 pixeis. resultante da multiplicação de 28x28 que é o formato das imagens na base de 
# dados. 
# activation: sigmoid que retorna valores entre 0 e 1 para cada saida(das 784). Pode-se utilizar a sigmoid porque
# se fez anteriormente a normalização dos pixeis.
# Camada reshape: Alterar o formato da camada de saida de vector para matriz 28x28(para se obter a imagem).

# L1L2 (Regularização)
# Adiciona uma penalização à funçao erro no treino do modelo para evitar o overfitting.
# Quando se faz o calculo do erro tem uma função adicional que penaliza o resultado se der um valor muito baixo.
# Normalmente é utilizado quando se tem muitos variaveis de entrada. O que acontece em imagens onde se tem um grande conjunto
# de pixeis(principalmente se a resolução for muito grande).

gerador = models.Sequential()
gerador.add(layers.Dense(units = 500, input_dim = 100, activation = 'relu', kernel_regularizer = regularizers.L1L2(1e-5, 1e-5)))
gerador.add(layers.Dense(units = 500, input_dim = 100, activation = 'relu', kernel_regularizer = regularizers.L1L2(1e-5, 1e-5)))
gerador.add(layers.Dense(units = 784, activation = 'sigmoid', kernel_regularizer = regularizers.L1L2(1e-5, 1e-5)))
gerador.add(layers.Reshape((28,28)))
```

## Modelo do Discriminador(Rede Neural Artificial)

```python
# Discriminador

# Camada de entrada: formato 28x28 das imagens do gerador e das imagens da base de dados X_train.
# Camada Flatten: transformar as imagens de matriz em vector para servir de entrada na rede neural densa.
# 1ª camada oculta: 500 neurónios na camada oculta.
# 2ª camada oculta: 500 neurónios na camada oculta.
# Camada de saida: units=1 porque é um problema de classificação binária. Pretende-se saber se a imagem corresponde ao 
# pretendido ou não e retornar um valor entre o e 1(uma probabilidade).
# activation = 'sigmoid', prque é a função utilizada para problemas de classificação binária.

# Pré-processamento das imagens
discriminador = models.Sequential()
discriminador.add(layers.InputLayer(input_shape=(28,28)))
discriminador.add(layers.Flatten())
# Rede neural artificial
discriminador.add(layers.Dense(units = 500, activation = 'relu', kernel_regularizer = regularizers.L1L2(1e-5, 1e-5)))
discriminador.add(layers.Dense(units = 500, activation = 'relu', kernel_regularizer = regularizers.L1L2(1e-5, 1e-5)))
discriminador.add(layers.Dense(units = 1, activation = 'sigmoid', kernel_regularizer = regularizers.L1L2(1e-5, 1e-5)))
```

## Compilação e Ajuste do modelo GAN

```python
# Modelo GAN
# Definir a rede neural gerador como a rede geradora de novas imagens.
# Definir a rede neural discriminador como a rede avaliadora das imagens geradas pelo gerador.
# Normal_latent_sampling: São gerados 100 valores aleatórios, amostras de uma distribuição normal, para formar uma imagem. Tem 
# de ser um valor igual ao input_dim do gerador. 

gan = simple_gan(gerador, discriminador, normal_latent_sampling((100,)))
```

```python
# Para o treino do modelo são utilizados os pesos da rede gerador e os pesos da rede discriminador
model = AdversarialModel(base_model = gan,
                          player_params = [gerador.trainable_weights, 
                                           discriminador.trainable_weights])
```

```python
# Compilação do modelo
# adversarial_optimizer: AdversarialOptimizerSimultaneous(), para actualizar as duas redes simultaneamente em cada um dos batch
# (batch a quatidade de registros utilizada para a actualização dos pesos).
# player_optimizers: Optimizador de cada uma das redes.

model.adversarial_compile(adversarial_optimizer = AdversarialOptimizerSimultaneous(),
                           player_optimizers = ['adam', 'adam'],
                           loss = 'binary_crossentropy')
```

```python
# Ajuste do modelo
# y: gan_targets(60000), sendo 60000 o número de imagens total de treino.
# batch_size: Quantas imagens passarão pelo gerador e pelo discriminador a cada batch. 

# Os pesos são atualizados depois que a última amostra do batch passa pela rede (essa é aliás a definição de batch), mais 
# especificamente logo depois que o gerador é avaliado em relação a sua capacidade de gerar imagens reais, ou seja, quando a 
# GAN inteira já teve todos os erros calculados (erro do discriminador e erro do gerador).   

model.fit(x = X_train, y = gan_targets(60000), epochs = 25, batch_size = 256)
```

```python
# gerar 20 imagens aleatorias
# 20 imagens com 100 dimensões(nº de dimensões da camada de entrada do gerador).
amostras = np.random.normal(size = (20,100))
```

```python
# imagens geradas: 20 imagens 28x18
# As imagens são criadas apenas utilizando o gerador. O processo de treino visou melhorar os pesos do gerador de modo que este
# receba numeros aleatorios e crie novas imagens melhores.
# Caso se pretendar gravar a rede neural, basta apenas gravar o gerador, porque depois do treino feito apenas é necessário
# o gerador.

previsao = gerador.predict(amostras)
```
